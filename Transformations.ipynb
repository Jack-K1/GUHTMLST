```python

# Read the CSV file
df = pd.read_csv('Combined_GU_trades.csv')

# Convert 'Open Time' and 'Close Time' to datetime format
df['Open Time'] = pd.to_datetime(df['Open Time'], format='%d.%m.%Y %H:%M')
df['Close Time'] = pd.to_datetime(df['Close Time'], format='%d.%m.%Y %H:%M')

# Sort the DataFrame by 'Open Time'
df = df.sort_values(by='Open Time')

# Reformat the date columns to 'yyyy-mm-dd'
df['Open Time'] = df['Open Time'].dt.strftime('%Y-%m-%d %H:%M')
df['Close Time'] = df['Close Time'].dt.strftime('%Y-%m-%d %H:%M')

# Save the sorted DataFrame back to a CSV file
df.to_csv('trades/finished_combined_trades.csv', index=False)

```

```python
import pandas as pd
def load_and_adjust_data(file_path):
    # Load the data
    data = pd.read_csv(file_path)
    
    # Combine the 'Date' and 'Time' columns into a single 'DateTime' column
    data['DateTime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'])
    
    # Identify rows on or after 2024-03-31
    shift_date = pd.to_datetime('2024-03-31')
    to_adjust = data['DateTime'] >= shift_date

    # Adjust the time by adding one hour for the relevant rows
    data.loc[to_adjust, 'DateTime'] = data.loc[to_adjust, 'DateTime'] + pd.Timedelta(hours=1)

    # Split the adjusted 'DateTime' back into 'Date' and 'Time'
    data['Date'] = data['DateTime'].dt.strftime('%Y%m%d')
    data['Time'] = data['DateTime'].dt.strftime('%H:%M:%S')

    # Drop the temporary 'DateTime' column
    data.drop(columns=['DateTime'], inplace=True)
    
    return data

# Adjust the file path as needed
file_path = '/workspaces/GUHTMLST/GUM1_OHLC_dropnaCSV.csv'
adjusted_data = load_and_adjust_data(file_path)

# Save the adjusted data to a new CSV file
adjusted_data.to_csv('GUM1_adjusted.csv', index=False)

# Display the adjusted data
print(adjusted_data.head(10))
```
